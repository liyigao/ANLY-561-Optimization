# Exercise 1  
1. $f(x)=|x|$  
Let $a, b\in \mathbb{R}$. $\forall t\in[0,1]$,
$$
f[(1-t)a+tb]=|(1-t)a+tb|
$$
$$
(1-t)f(a)+tf(b)=(1-t)|a|+t|b|=|(1-t)a|+|tb|
$$
By Triangle Inequality, $|(1-t)a+tb|\leq |(1-t)a|+|tb|$  
Therefore, $f[(1-t)a+tb]\leq (1-t)f(a)+tf(b)$  
$f(x)=|x|$ is convex on $\mathbb{R}$
2. $f(x)=x^2$  
$$
f'(x)=2x
$$
$$
f''(x)=2
$$
$\forall x\in\mathbb{R}$, let $c=1>0$
$$
f''(x)>c, \forall x\in\mathbb{R}
$$
$f(x)=x^2$  is strongly convex on $\mathbb{R}$
3. $f(x)=x^3-x, x\in[0,1]$  
$$
f'(x)=3x^2-1
$$
$$
f''(x)=6x
$$
$\forall x\in [0,1]$, $f''(x)\geq 0$. By Second Order Conditions for Convexity, $f(x)=x^3-x$ is strictly convex on $[0,1]$
4. $f(x)=|x-\mu|$, for fixed $\mu\in\mathbb{R}$  
Let $a, b\in \mathbb{R}$. $\forall t\in[0,1]$,
$$
f[(1-t)a+tb]=|(1-t)a+tb-\mu|
$$
$$
(1-t)f(a)+tf(b)=(1-t)|a-\mu|+t|b-\mu|=|(1-t)a-(1-t)\mu|+|tb-t\mu|
$$
By Triangle Inequality, $|(1-t)a-(1-t)\mu|+|tb-t\mu|\geq|(1-t)a+tb-(1-t)\mu-t\mu|=|(1-t)a+tb-\mu|$  
Therefore, $f[(1-t)a+tb]\leq (1-t)f(a)+tf(b)$  
$f(x)=|x-\mu|$, for fixed $\mu\in\mathbb{R}$ is convex on $\mathbb{R}$
5. $f(x)=\frac{(x-\mu)^2}{2\sigma^2}$, for fixed $\mu\in\mathbb{R}$ and $\sigma>0$  
$$
f'(x)=\frac{x-\mu}{\sigma^2}
$$
$$
f''(x)=\frac{1}{\sigma^2}
$$
Take $c=\frac{1}{2\sigma^2}$, $f''(x)>c$ for all $x\in\mathbb{R}$  
Therefore, $f(x)=\frac{(x-\mu)^2}{2\sigma^2}$, for fixed $\mu\in\mathbb{R}$ and $\sigma>0$ is strongly convex on $x\in\mathbb{R}$
6. $f(x)=\sum_{i=1}^n|x_i-x|$ where $x_1$,...,$x_n\in\mathbb{R}$ are fixed values  
From part (4d), we proved that $f(x)=|x_i-x|$ is convex on $\mathbb{R}$ for each $x_i$  
By theorem, sum of convex functions is also convex.  
Thus, $f(x)=\sum_{i=1}^n|x_i-x|$ is also convex on $\mathbb{R}$.
7. $f(x)=\frac{(x-\mu)^2}{2\sigma^2}+\lambda|x|$ for fixed values $\mu\in\mathbb{R}$, $\sigma>0$ and $\lambda>0$  
From part (1a) and (5e), we proved that $f(x)=|x|$ and $f(x)=\frac{(x-\mu)^2}{2\sigma^2}$ are convex on $\mathbb{R}$  
By theorem, positive weighted sum of convex functions is also convex.  
Thus $f(x)=\frac{(x-\mu)^2}{2\sigma^2}+\lambda|x|$ is also convex on $\mathbb{R}$  


# Exercise 2  
1. $f(x)=|x|$  
$$
f(x)=|x|=\begin{cases} x & x\geq0 \\ -x & x<0 \end{cases}
$$
When $x\geq0$, f(x) is monotone, increasing function. So $\min_{x\geq0}f(x)=f(0)=0$  
When $x<0$, f(x) is monotone, decreasing function. So $\min_{x<0}f(x)=f(0^-)=-0^-=0^+$  
Overall, $\min_{x\in\mathbb{R}}f(x)=f(0)=0$
2. $f(x)=x^2$  
$$
f'(x)=2x=0\implies x=0
$$
Therefore, $\min_{x\in\mathbb{R}}f(x)=f(0)=0^2=0$
3. $f(x)=x^3-x, x\in[0,1]$  
$$
f'(x)=3x^2-1=0\implies x=\frac{1}{\sqrt{3}}
$$
Therefore, $\min_{x\in[0,1]}f(x)=f(\frac{1}{\sqrt{3}})=-\frac{2}{3\sqrt{3}}$
4. $f(x)=|x-\mu|$, for fixed $\mu\in\mathbb{R}$  
$$
f(x)=|x-\mu|=\begin{cases} x-\mu & x\geq\mu \\ \mu-x & x<\mu \end{cases}
$$
When $x\geq\mu$, f(x) is monotone, increasing function. So $\min_{x\geq\mu}f(x)=f(\mu)=\mu-\mu=0$  
When $x<\mu$, f(x) is monotone, decreasing function. So $\min_{x<\mu}f(x)=f(\mu^-)=\mu-\mu^-=0^+$  
Overall, $\min_{x\in\mathbb{R}}f(x)=f(\mu)=0$
5. $f(x)=\frac{(x-\mu)^2}{2\sigma^2}$, for fixed $\mu\in\mathbb{R}$ and $\sigma>0$  
$$
f'(x)=\frac{x-\mu}{\sigma^2}=0 \implies x=\mu
$$
Therefore, $\min_{x\in\mathbb{R}}f(x)=f(\mu)=0$
6. $f(x)=\sum_{i=1}^n|x_i-x|$ where $x_1$,...,$x_n\in\mathbb{R}$ are fixed values  
Without loss of generality (WLOG), assume that $x_1\leq x_2\leq ...\leq x_n$.  
When $n=2k-1$ where $k\in\mathbb{Z}^+$, $\min_{x\in\mathbb{R}}f(x)=f(x_k)=\sum_{i=k+1}^n x_i - \sum_{i=1}^{k-1} x_i$  
When $n=2k$ where $k\in\mathbb{Z}^+$, $f(x)$ reaches minimum when $x\in[x_k,x_{k+1}]$. Let $x_k'$ be any number in $[x_k,x_{k+1}]$. $\min_{x\in\mathbb{R}}f(x)=f(x_k')=\sum_{i=k+1}^n x_i - \sum_{i=1}^k x_i$
7. $f(x)=\frac{(x-\mu)^2}{2\sigma^2}+\lambda|x|$ for fixed values $\mu\in\mathbb{R}$, $\sigma>0$ and $\lambda>0$  
When $\mu>0$, $f(x)$ reaches minimum only at $x>0$:  
$$
f'(x)=\frac{x-\mu}{\sigma^2}+\lambda=0 \implies x=\mu-\lambda\sigma^2
$$
So $\min_{x\in\mathbb{R}}f(x)=f(\mu-\lambda\sigma^2)=\lambda\mu-\frac{\lambda^2\sigma^2}{2}$  
When $\mu<0$, $f(x)$ reaches minimum only at $x>0$:  
$$
f'(x)=\frac{x-\mu}{\sigma^2}-\lambda=0 \implies x=\mu+\lambda\sigma^2
$$
So $\min_{x\in\mathbb{R}}f(x)=f(\mu+\lambda\sigma^2)=-\lambda\mu-\frac{\lambda^2\sigma^2}{2}$  
When $\mu=0$, $\min_{x\in\mathbb{R}}f(x)=f(0)=0$
Overall, $\min_{x\in\mathbb{R}}f(x)=\lambda|\mu|-\frac{\lambda^2\sigma^2}{2}$  


# Exercise 3  
1. $f(x)=|x|$  
$$
f'(x)=\begin{cases} 1 & x\geq0 \\ -1 & x\leq0 \end{cases}
$$
$$
\max_{x\in\mathbb{R}}|f'(x)|=1=C<\infty
$$
Therefore, $f(x)=|x|$ is Lipschitz on $\mathbb{R}$ and Lipschitz constant $C=1$.
2. $f(x)=x^2$
$$
f'(x)=2x
$$
In this case, $\max_{x\in\mathbb{R}}|f'(x)|$ does not exist on $\mathbb{R}$ ($\max\rightarrow\infty$). So $f(x)=x^2$ is not Lipschitz on $\mathbb{R}$.
3. $f(x)=x^3-x, x\in[0,1]$
$$
f'(x)=3x^2-1
$$
$$
\max_{x\in[0,1]}|f'(x)|=|f(1)|=|3\times1^2-1|=2=C<\infty
$$
Therefore, $f(x)=x^3-x$ is Lipschitz on $[0,1]$ and Lipschitz constant $C=2$.
4. $f(x)=|x-\mu|$, for fixed $\mu\in\mathbb{R}$  
$$
f'(x)=\begin{cases} 1 & x\geq\mu \\ -1 & x\leq\mu \end{cases}
$$
$$
\max_{x\in\mathbb{R}}|f'(x)|=1=C<\infty
$$
Therefore, $f(x)=|x-\mu|$ is Lipschitz on $\mathbb{R}$ and Lipschitz constant $C=1$.
5. $f(x)=\frac{(x-\mu)^2}{2\sigma^2}$, for fixed $\mu\in\mathbb{R}$ and $\sigma>0$  
$$
f'(x)=\frac{x-\mu}{\sigma^2}
$$
In this case, $\max_{x\in\mathbb{R}}|f'(x)|$ does not exist on $\mathbb{R}$ ($\max\rightarrow\infty$). So $f(x)=\frac{(x-\mu)^2}{2\sigma^2}$ is not Lipschitz on $\mathbb{R}$.
6. $f(x)=\sum_{i=1}^n|x_i-x|$ where $x_1$,...,$x_n\in\mathbb{R}$ are fixed values
Without loss of generality (WLOG), assume that $x_1\leq x_2\leq ...\leq x_n$.  
$$
f'(x)=\sum_{i=1}^n sgn(x_i-x)=-n,-n+1,...-1,0,1,...,n-1,n
$$
$$
\max_{x\in\mathbb{R}}|f'(x)|=n=C<infty
$$
Therefore, $f(x)=\sum_{i=1}^n|x_i-x|$ is Lipschitz on $\mathbb{R}$ and Lipschitz constant $C=n$.
7. $f(x)=\frac{(x-\mu)^2}{2\sigma^2}+\lambda|x|$ for fixed values $\mu\in\mathbb{R}$, $\sigma>0$ and $\lambda>0$  
$$
f'(x)=\begin{cases} \frac{x-\mu}{\sigma^2} + \lambda & x\geq0 \\ \frac{x-\mu}{\sigma^2} - \lambda & x\leq0 \end{cases}
$$
In this case, $|f'(x)|$ is not bounded above. When $x\rightarrow\infty$, $f'(x)\rightarrow\infty$. Thus, $f(x)=\frac{(x-\mu)^2}{2\sigma^2}+\lambda|x|$ is not Lipschitz on $\mathbb{R}$.
