(a) Convert maximization problem $max_{{\bf x}\in\mathbb{R}^2}x_1\text{ subject to }x_2+(x_1-1)^3\leq0, x_1\geq0, x_2\geq0$ into a minimization problem:  
$$min_{{\bf x}\in\mathbb{R}^2}-x_1\text{ subject to }x_2+(x_1-1)^3\leq0, x_1\geq0, x_2\geq0$$
$$f({\bf x})=-x_1, h_1({\bf x})=x_2+(x_1-1)^3, h_2({\bf x})=-x_1, h_3({\bf x})=-x_2$$
$$\nabla h_1({\bf x})=\begin{pmatrix} 3(x_1-1)^2 \\ 1 \end{pmatrix}, \nabla h_2({\bf x})=\begin{pmatrix} -1 \\ 0 \end{pmatrix}, \nabla h_3({\bf x})=\begin{pmatrix} 0 \\ -1 \end{pmatrix}$$  
Assume that ${\bf x}^*=\begin{pmatrix} 1 \\ 0 \end{pmatrix}$  
$$\nabla h_1({\bf x}^*)=\begin{pmatrix} 0 \\ 1 \end{pmatrix}, \nabla h_2({\bf x}^*)=\begin{pmatrix} -1 \\ 0 \end{pmatrix}, \nabla h_3({\bf x}^*)=\begin{pmatrix} 0 \\ -1 \end{pmatrix}$$  
Because $\nabla h_1({\bf x}^*)=-1\times\nabla h_3({\bf x}^*)$, $\nabla h_1({\bf x}^*)$ and $\nabla h_3({\bf x}^*)$ are linearly dependent,  ${\bf x}^*=\begin{pmatrix} 1 \\ 0 \end{pmatrix}$ fails Linear Independence Constraint Qualification.  
  
(b) $\nabla f({\bf x})=\begin{pmatrix} -1 \\ 0 \end{pmatrix}$  
KKT conditions:  
Stationarity:  
$$\begin{pmatrix}-1\\0\end{pmatrix}+\mu_1\begin{pmatrix}3(x_1-1)^2\\1\end{pmatrix}+\mu_2\begin{pmatrix}-1\\0\end{pmatrix}+\mu_3\begin{pmatrix}0\\-1\end{pmatrix}={\bf 0}$$
Primal Feasibility:  
$$h_1({\bf x})=x_2+(x_1-1)^3\leq0,h_2({\bf x})=-x_1\leq0,h_3({\bf x})=-x_2\leq0$$
Dual Feasibility:  
$$\mu_1,\mu_2,\mu_3\geq0$$
Complementary Slackness:  
$$\mu_1(x_2+(x_1-1)^3)=\mu_2(-x_1)=\mu_3(-x_2)=0$$

(c) Again, assume that ${\bf x}^*=\begin{pmatrix} 1 \\ 0 \end{pmatrix}$ is the optimal solution.  
From Complementary Slackness, $\mu_2=0$. Applying $\mu_2$ to Stationarity:  
$$\begin{pmatrix}-1\\0\end{pmatrix}+\mu_1\begin{pmatrix}0\\1\end{pmatrix}+{\bf 0}+\mu_3\begin{pmatrix}0\\-1\end{pmatrix}={\bf 0}$$
$$\implies-1=0\text{ Contradiction}$$  
So, ${\bf x}^*=\begin{pmatrix} 1 \\ 0 \end{pmatrix}$ does not satisfy KKT conditions.  
  
(d) $x_2\leq-(x_1-1)^3, x_1\geq0, x_2\geq0$  

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.,2.,0.01)
s = np.arange(0.,1.,0.01)
plt.plot(t, -(t-1)**3)
plt.grid(True)
plt.show()
plt.plot(s, -(s-1)**3)
plt.grid(True)
plt.show()
```

For the region $x_2\leq-(x_1-1)^3$ in the first quadrant, in order to maximize $x_1$, the only solution of this maximization problem is $(1,0)$.

$$h_2({\bf x})=x_1-1\leq0$$
$$h_3({\bf x})=-x_2\leq0$$
$$h_4({\bf x})=x_2-1\leq0$$
$$h_5({\bf x})=-x_3\leq0$$
$$h_6({\bf x})=x_3-1\leq0$$
